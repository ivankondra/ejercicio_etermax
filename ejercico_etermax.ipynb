```python
import pandas as pd
from google.colab import drive
import os

drive.mount('/content/drive')

folder_path = "/content/drive/MyDrive/csv_etermax/corpus/"

```

```python
#armo un df a partir de los valores de los csvs

second_line_values = []
archivo = []

for file_name in os.listdir(folder_path):
    archivo.append(file_name) #mantengo el nombre del archivo original para no perder esa info
    if file_name.endswith(".csv"):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path) as f:
            second_line = f.readlines()[1].strip()  
            second_line_values.append(second_line)

#armo df a partir de un diccionario
df = pd.DataFrame({"Pregunta": second_line_values, "Archivo": archivo})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from nltk.corpus import stopwords
stop_words_sp = stopwords.words('spanish')


# Create a TF-IDF vectorizer object
vectorizer = TfidfVectorizer(stop_words=stop_words_sp)

# Create a TF-IDF matrix from the text column
tfidf_matrix = vectorizer.fit_transform(df["Pregunta"])

# Train a KMeans clustering model on the TF-IDF matrix
num_clusters = 10  # Change this to the number of desired clusters
km = KMeans(n_clusters=num_clusters)
km.fit(tfidf_matrix)

# Add a new column to the DataFrame with the cluster labels
df["topico"] = km.labels_

# Print the number of rows in each cluster
print(df["topico"].value_counts())
```
